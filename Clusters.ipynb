c)Supondo que você queira tentar diferentes valores de K (por exemplo, 2, 3, 4), calcule WCSS para diferentes valores de K

```r
cd <- numeric(length = 10) # Considerando valores de K de 1 a 10

for (k in 1:10) {
  kr <- kmeans(dados[, 1:7], centers = k)
  cd[k] <- sum(kr$withinss)
}
```

```r
plot(1:10, cd, type = "b", pch = 19, frame = FALSE,
     xlab = "Número K de cluster", ylab = "Soma de quadrados dentro do cluster",
     main = "Gráfico de cotovelo")
abline(v = which.min(-diff(cd)), col = "blue", lty = 1)
```
